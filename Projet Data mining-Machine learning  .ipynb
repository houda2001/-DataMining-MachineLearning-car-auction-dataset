# Performing mean value imputation for all numerical features and frequent occurring value for categorical features. 

```python
Numerical=["MMRCurrentRetailCleanPrice","MMRCurrentRetailAveragePrice","MMRCurrentAuctionCleanPrice","MMRCurrentAuctionAveragePrice","MMRAcquisitonRetailCleanPrice",
"MMRAcquisitionRetailAveragePrice","MMRAcquisitionAuctionCleanPrice","MMRAcquisitionAuctionAveragePrice"]
categorical=["Size","WheelType","WheelTypeID","Transmission","Color"]
for c in Numerical:
    dataset.loc[dataset[c].isna(),c] = dataset[c].mean()
for c in categorical:
    dataset.loc[dataset[c].isna(),c] = dataset[c].value_counts().sort_values(ascending=False).index[0]   
```

```python
# Make sure that we don't have any missing value 
dataset.isna().sum()
```

# DNN

Import tensorflow and classification_report

```python
import tensorflow as tf  # library, allowing to run machine learning and deep learning applications
from sklearn.metrics import classification_report # to show Text summary of the precision, recall, F1 score for each class
```

```python
# Defining the columns that our neural network will train from 
def create_feature_column():  
    
    data=dataset.drop(columns=['IsBadBuy']).columns.tolist()
    i=0
    feat=[]
    for item in data:
        feat.append(tf.feature_column.numeric_column(item))
        
    feature_column=feat
    return feature_column
feature_column = create_feature_column()
```

```python
#Define the data process 

def Data_Process(dataset):
    df=dataset
    print(df.head(10))
    X_Data=dataset.drop(columns=['IsBadBuy'])
    Y_Data=dataset['IsBadBuy']
    X_Train , X_Test , Y_Train , Y_Test= train_test_split(X_Data,Y_Data,test_size=0.3,random_state=101)
    return X_Train, X_Test, Y_Train, Y_Test
X_Train , X_Test , Y_Train , Y_Test = Data_Process(dataset)  

```

```python
# Define the dnn model 
dnnmodel=tf.compat.v1.estimator.DNNClassifier(hidden_units =[512,256,128],   
                                             feature_columns=feature_column,
                                             n_classes=2,
                                             activation_fn=tf.nn.softmax,
                                             dropout=None,
                                             optimizer=tf.compat.v1.train.AdamOptimizer(learning_rate=0.01))
```

```python
# creation of input data according to the Tensorflow model

input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(
    X_Train,
    Y_Train,
    batch_size=50,
    num_epochs=1000,
    shuffle=True)
dnnmodel.train(input_fn=input_func, steps=1000)
```

```python
#creation of evaluation data according to the tensorflow model

eval_func = tf.compat.v1.estimator.inputs.pandas_input_fn(X_Test,
                                               Y_Test,
                                               batch_size=50,
                                               num_epochs=1,
                                               shuffle=False)
 
dnnmodel.evaluate(eval_func)                                                                                              
```

```python
# prediction of the target  

predict_input_fn = tf.compat.v1.estimator.inputs.pandas_input_fn(
      x=X_Test,
      num_epochs=1,
      shuffle=False)

predictions = list(dnnmodel.predict(input_fn=predict_input_fn))
prediction = [p["class_ids"][0] for p in predictions]

# classification_report
CR = classification_report(Y_Test,prediction)

#confusion matrix 

CM = confusion_matrix(Y_Test,prediction)
```
